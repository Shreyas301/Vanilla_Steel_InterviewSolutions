```python
import pandas as pd
import numpy as np
```

```python
supplier1 = pd.read_excel("supplier_data_1.xlsx")
supplier2 = pd.read_excel("supplier_data_2.xlsx")
```

```python
print(supplier1.head())
print(supplier2.head())
```

```python
supplier1.columns = supplier1.columns.str.lower().str.replace(" ", "_")
supplier2.columns = supplier2.columns.str.lower().str.replace(" ", "_")
```

```python
supplier1.fillna(method='ffill', inplace=True)  # Forward-fill missing values
supplier2.dropna(subset=['material_name'], inplace=True)  # Drop rows where 'material' is missing
```

```python
supplier1_cleaned = supplier1.dropna()
supplier2_cleaned = supplier2.dropna()
```

```python
print(supplier1_cleaned.isnull().sum())
print(supplier2_cleaned.isnull().sum())
```

```python
print(supplier1_cleaned.head())
print(supplier2_cleaned.head())
```

```python
supplier1_cleaned.dtypes
supplier2_cleaned.dtypes
```

```python
supplier1_cleaned.to_excel("supplier_cleaned_file1.xlsx", index=False)
supplier2_cleaned.to_excel("supplier_cleaned_file2.xlsx", index=False)
```

```python
inventory_dataset = pd.concat([supplier1_cleaned, supplier2_cleaned], ignore_index=True)
inventory_dataset.to_csv("inventory_dataset.csv", index=False)
```
